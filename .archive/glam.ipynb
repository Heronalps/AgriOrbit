```python
import requests

base_url = "https://s9h5qqvw16.execute-api.us-east-1.amazonaws.com/dev/"
product_id = "mod09a1-ndwi"  # This is correct as per your original request
date = "2022-09-21"
lon = -76.937759
lat = 38.989697

# Construct the URL
url = f"{base_url}point/{product_id}/{date}/{lon}/{lat}/"

# Optional query parameters
params = {
    "cropmask_id": "no-mask",  # You can change this as needed
    "anomaly": "full",         # Optional: '5year', '10year', or 'full'
    "anomaly_type": "mean",    # Optional: 'mean', 'median', or 'diff'
    # "diff_year": 2021,       # Optional: Include if you want to see difference from a specific year
}

# Make the request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print("Point data:", data)
else:
    print(f"Error: {response.status_code}")
    print(response.text)
```
